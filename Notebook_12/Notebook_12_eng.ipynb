```python
# Statistic measurements
print(df['Fare'].mean())
print(df['Fare'].max())
print(df['Fare'].min())
print(df['Fare'].median())
```

```python
df["Name"].str.lower() # string operations
```

```python
# It is possible to convert one column of the DataFrame to the Series
s = df["Fare"]
type(s)
```

```python
print(s.mean())
print(s.max())
print(s.min())
print(s.median())
```
