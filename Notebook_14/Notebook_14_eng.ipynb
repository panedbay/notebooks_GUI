# Notebook_14 - Visualization with pandas

Now that we hav sufficient knowledge about matplotlib, let us try to plot the data that is given by pandas DataFrame.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../tested.csv")
df
```

```python
df = df[df["Fare"] < 500] # exclude fares > 500
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Fare"], df["Age"], 'o')
plt.show()
```

Also there are pandas functions that allow for plotting.

```python
plt.bar(df["Pclass"], df["Fare"])
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

```python
plt.pie(df.iloc[0:10]["Fare"], labels=df.iloc[0:10]["Fare"])
plt.show()
```

```python
plt.plot(df["Age"], linestyle = 'dotted')
plt.show()
```

```python
plt.bar(df["Sex"], df["Age"])
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
new_df = df[df["Age"] <= 18]
new_df.head(10)
```

```python
plt.bar(new_df["Pclass"], new_df["Fare"])
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

```python
plt.plot(new_df["Fare"], new_df["Age"], 'o')
plt.show()
```

```python
plt.hist(df["Embarked"], bins=3)
plt.show()
```
